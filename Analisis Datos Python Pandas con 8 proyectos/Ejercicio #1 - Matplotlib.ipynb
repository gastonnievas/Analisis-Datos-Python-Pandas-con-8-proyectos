1- Vamos a hacer un gráfico simple que representa el gasto de alguien en el almuerzo durante la semana pasada. Primero, defina dos listas, `days` y `money_spent`, que contengan los siguientes números enteros:



```python
from matplotlib import pyplot as plt
```

```python
days = [0,1,2,3,4,5,6]
money_spent = [10,12,12,10,14,22,24]
```

2- Graficar `days` en el eje x y `money_spent` en el eje y usando `plt.plot()`. Mostrar el gráfico usando `plt.show()`.

```python
plt.plot(days, money_spent)
plt.show()
```

3- Hemos definido listas llamadas `time`, `revenue` y `costs`. 

```python
time = [0, 1, 2, 3, 4]
revenue = [200, 400, 650, 800, 850]
costs = [150, 500, 550, 550, 560]
```

4- Graficar el `revenue` vs. `time`.

```python
plt.plot(time, revenue)
plt.show()
```

5- Graficar `costs` vs `time` en el mismo grafico del `time` vs `revenue`

```python
plt.plot(time, costs)
plt.show()
```

6- Grafique `revenue` vs `time` como una linea purple (`purple`), y dashed (`--`).

```python
plt.plot(time, revenue, color='purple', linestyle = '--')
plt.show()
```

7- Grafique `costs` vs `times` como una linea con color HEX `#82edc9` y marcadores cuadrados (`s`)

```python
plt.plot(time, costs, color ='#82edc9', marker = 's')
plt.show()
```

8- Graficque los dos anteriores en un mismo grafico

```python
plt.plot(time, revenue, color='purple', linestyle = '--')
plt.plot(time, costs, color ='#82edc9', marker = 's')
plt.show()
```

9- Hemos trazado una línea que representa el gasto de alguien en café durante los últimos 12 años. Corra el codigo para ver el resultado

```python
x = range(12)
y = [3000, 3005, 3010, 2900, 2950, 3050, 3000, 3100, 2980, 2980, 2920, 3010]
plt.plot(x, y)
```

10- Modifiquemos los ejes para ampliar un poco más nuestro gráfico de líneas. Utilice `plt.axis()` para modificar los ejes para que el eje x vaya de 0 a 12, y el eje y vaya de 2900 a 3100.

```python
plt.plot(x, y)
plt.axis([0, 12, 2900, 3100])
plt.show()
```

11 - Marca el ejex como `Time`

```python
plt.plot(x,y)
plt.axis([0, 12, 2900, 3100])
plt.xlabel('Time')
plt.show()
```

12- Marca el eje-y como `Dollars spent on coffee`

```python
plt.plot(x,y)
plt.axis([0, 12, 2900, 3100])
plt.xlabel('Time')
plt.ylabel('Dollars spent on coffee')
plt.show()
```

13- Añade el titulo `My Last Twelve Years of Coffee Drinking`

```python
plt.plot(x,y)
plt.axis([0, 12, 2900, 3100])
plt.xlabel('Time')
plt.ylabel('Dollars spent on coffee')
plt.title('My Last Twelve Years of Coffee Drinking')
plt.show()
```
