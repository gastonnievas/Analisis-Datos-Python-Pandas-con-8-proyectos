```python
sales = pd.DataFrame({
    'month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'revenue': [300, 290, 310, 325, 475, 495]
})

targets = pd.DataFrame({
    'month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'target': [310, 270, 300, 350, 475, 500]
})
```

```python
sales_vs_targets = pd.merge(sales, targets)
sales_vs_targets
```

```python
lo_logramos = sales_vs_targets[sales_vs_targets.revenue > sales_vs_targets.target]
lo_logramos
```

```python
big_df = orders.merge(customers).merge(products)
big_df
```

```python
men_women = pd.DataFrame({
    'month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'men': [30, 29, 31, 32, 47, 49],
    'women': [35, 35, 29, 28, 50, 45]
})
```

```python
all_data = sales.merge(targets).merge(men_women)
all_data
```

```python
## Seleccionamos las filas donde revenue es mayor a target
## y donde women es mayor a men. 
results = all_data[(all_data.revenue > all_data.target) & (all_data.women > all_data.men)]
results
```

```python
store_a = pd.read_csv('datasets/store_a.csv')
store_b = pd.read_csv('datasets/store_b.csv')
print(store_a)
print(store_b)
```

```python
# OUTER, INNER
store_a_b_outer = pd.merge(store_a, store_b, how='outer')
# El outer completara con NaN los que no tienen dato. 
# Podemos reemplazar los NaN por ceros para poder hacer calculos luego, usando fillna(0)
store_a_b_outer.fillna(0, inplace=True)
store_a_b_outer
```

```python
#LEFT MERGE
pd.merge(store_a, store_b, how='left')
# Trae todos los datos de la tabla izquierda (store_a)
```

```python
# RIGHT es lo inverso a LEFT. Generalmente usamos LEFT y cambiamos el orden de a y b
pd.merge(store_a, store_b, how='right')
```
